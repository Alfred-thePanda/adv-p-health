```python
y = df['DIABETE4']
X = df.drop(['DIABETE4'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
from flaml import AutoML
automl = AutoML()
automl.fit(X_train, y_train, task="classification", time_budget=300)
```

```python
# Print the best model
print(automl.model.estimator)
```

```python
import pickle
with open('./models/diabetes_automl_2.pkl', 'wb') as f:
    pickle.dump(automl, f, pickle.HIGHEST_PROTOCOL)
```

```python
from sklearn.metrics import accuracy_score
y_pred = automl.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
for i in range(0,len(automl.feature_names_in_)):
    print('{:32}: {}\t{}'.format(automl.feature_names_in_[i], (automl.feature_importances_[i]*146).round(2), df.columns[i]))
```

```python
for f in automl.feature_importances_:
    print('{}, '.format((f*146).round(2)))
```

```python
import matplotlib.pyplot as plt
#plt.figure().set_figwidth(25)
#plt.figure().set_figheight(20)
plt.rcParams['figure.figsize'] = [15, 39]
plt.barh(automl.feature_names_in_, automl.feature_importances_)
plt.show()
```

```python
df2 = df.copy()
```

```python
df2['heart_prob'] = df[['CVDSTRK3', 'CVDCRHD4', 'CVDINFR4']].apply(lambda row: row.CVDSTRK3 or row.CVDCRHD4 or row.CVDINFR4, axis=1)
df2 = df2.drop(['CVDSTRK3', 'CVDCRHD4', 'CVDINFR4'],axis=1)
df2['heart_prob'].describe()
```

```python
y2 = df2['heart_prob']
X2 = df2.drop(['heart_prob'],axis=1)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=0)
```

```python
# undersample the over represented class without heart problems (90% of training set down to equal to 10%)
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
under_sampler = RandomUnderSampler()
X_res, y_res = under_sampler.fit_resample(X_train2, y_train2)
print(f"Training target statistics: {Counter(y_res)}")
print(f"Testing target statistics: {Counter(y_test2)}")
```

```python
heartml = AutoML()
heartml.fit(X_res, y_res, task="classification", time_budget=20)
```

```python
# testing heart problem accuracy
y_pred2 = heartml.predict(X_test2)
accuracy_score(y_test2, y_pred2)
```

```python
import matplotlib.pyplot as plt
#plt.figure().set_figwidth(25)
#plt.figure().set_figheight(20)
plt.rcParams['figure.figsize'] = [15, 39]
plt.barh(heartml.feature_names_in_, heartml.feature_importances_)
plt.show()
```

```python
# Print the best model
print(heartml.model.estimator)
```

```python
with open('./models/heart_automl.pkl', 'wb') as f:
    pickle.dump(heartml, f, pickle.HIGHEST_PROTOCOL)
```
